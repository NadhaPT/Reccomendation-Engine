```python
#popularity recommender system

def TopNPopularMovies(genre, threshold, topN):
    popularity = merged_data.groupby(['genres', 'title']).agg({'rating':['mean', 'size']}).reset_index()
    popularity.columns=['Genres', 'Title', 'Average_Ratings', 'Number_of_Ratings']
    
    #filter the data
    topNrecommendations = popularity[(popularity['Genres']==genre)&(popularity['Number_of_Ratings']>=threshold)]
    
    #output
    topNrecommendations['Sno.'] = range(1, len(topNrecommendations)+1)
    topNrecommendations.index = range(0, len(topNrecommendations))
    topNrecommendations.columns = ['Genres', 'Movie Title', 'Average Movie Rating', 'Number of Reviews', 'Sno.']
    return topNrecommendations[['Sno.', 'Movie Title', 'Average Movie Rating', 'Number of Reviews']]


```

```python
import pandas as pd
import numpy as np

def recommendation_genre(movie_df, similarity_matrix, movie_title, topN):
    # Create a mapping from movie titles to their indices
    indices = pd.Series(movie_df.index, index=movie_df['title']).to_dict()

    # Check if the movie title exists in the indices
    if movie_title not in indices:
        raise ValueError(f"Movie '{movie_title}' not found in the dataset.")

    # Index of the target movie
    index = indices[movie_title]

    # Cosine similarity scores
    cosine_scores = list(enumerate(similarity_matrix[index]))
    cosine_scores = sorted(cosine_scores, key=lambda x: x[1], reverse=True)[1:topN+2]

    # Extract the matching movies
    matched = [i[0] for i in cosine_scores]
    matching_df = movie_df.iloc[matched]

    # Filter out the target movie
    matching_df = matching_df[matching_df['title'] != movie_title]

    # Output
    matching_df.rename(columns={'title': 'Movie Title'}, inplace=True)
    matching_df['Sno.'] = range(1, len(matching_df) + 1)
    matching_df.index = range(0, len(matching_df))

    return matching_df[['Sno.', 'Movie Title']].head(topN)

# Test case
movies3 = pd.DataFrame({
    'title': ['Toy Story (1995)', 'Jumanji (1995)', 'Grumpier Old Men (1995)', 'Heat (1995)', 'Sabrina (1995)'],
    'genres': ['Adventure|Animation|Children|Comedy|Fantasy', 'Adventure|Children|Fantasy', 'Comedy|Romance', 'Action|Crime|Thriller', 'Comedy|Romance']
})

# Simulate a similarity matrix for demonstration purposes
cosine_sim = np.random.rand(len(movies3), len(movies3))

# Call the function
recommended_movies = recommendation_genre(movie_df=movies3, similarity_matrix=cosine_sim, movie_title='Toy Story (1995)', topN=2)
print(recommended_movies)
```

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Define the recommendation function
def recommendation_genre(movie_df, similarity_matrix, movie_title, topN):
    indices = pd.Series(movie_df.index, index=movie_df['title']).to_dict()
    if movie_title not in indices:
        raise ValueError(f"Movie '{movie_title}' not found in the dataset.")
    index = indices[movie_title]
    cosine_scores = list(enumerate(similarity_matrix[index]))
    cosine_scores = sorted(cosine_scores, key=lambda x: x[1], reverse=True)[1:topN+2]
    matched = [i[0] for i in cosine_scores]
    matching_df = movie_df.iloc[matched]
    matching_df = matching_df[matching_df['title'] != movie_title]
    matching_df.rename(columns={'title': 'Movie Title'}, inplace=True)
    matching_df['Sno.'] = range(1, len(matching_df) + 1)
    matching_df.index = range(0, len(matching_df))
    return matching_df[['Sno.', 'Movie Title', 'average_rating', 'num_reviews']].head(topN)


movies3= pd.DataFrame({
    'title': ['Toy Story (1995)', 'Jumanji (1995)', 'Grumpier Old Men (1995)', 'Heat (1995)', 'Sabrina (1995)'],
    'genres': ['Adventure|Animation|Children|Comedy|Fantasy', 'Adventure|Children|Fantasy', 'Comedy|Romance', 'Action|Crime|Thriller', 'Comedy|Romance'],
    'average_rating': [4.5, 4.0, 3.5, 4.2, 3.8],
    'num_reviews': [100, 150, 200, 175, 90]
})

# Simulated similarity matrix
cosine_sim = np.random.rand(len(movies3), len(movies3))

# Interactive Widgets
# Popularity Inputs
genres = widgets.Dropdown(options=list(set(movies3['genres'])), description="Genres", style={"description_width": "initial"})
num_reviews = widgets.IntText(description="Minimum Reviews", style={"description_width": "initial"})
num_recommendations_1 = widgets.IntText(description="Number of Recommendations", style={"description_width": "initial"})

# Tabs
b1 = widgets.Button(description="RECOMMEND ME", style={"description_width": "initial"})
h1 = widgets.HBox([num_reviews, num_recommendations_1])
popularity_tab = widgets.VBox([genres, h1, b1])

# Content-Based Inputs
title = widgets.Textarea(description="Movie Title", style={"description_width": "initial"})
num_recommendations_2 = widgets.IntText(description="Number of Recommendations", style={"description_width": "initial"})

# Tabs
b2 = widgets.Button(description="RECOMMEND ME", style={"description_width": "initial"})
h2 = widgets.HBox([num_recommendations_2])
content_tab = widgets.VBox([title, h2, b2])

# Display Tabs
tab = widgets.Tab(children=[popularity_tab, content_tab])
tab.set_title(0, 'Popularity Based Rec')
tab.set_title(1, 'Content Based Rec')
display(tab)

# Define the callbacks
def on_popularity_recommendation_clicked(b):
    genre_selected = genres.value
    min_reviews = num_reviews.value
    topN = num_recommendations_1.value
    
    filtered_movies = movies3[(movies3['genres'] == genre_selected) & (movies3['num_reviews'] >= min_reviews)]
    filtered_movies = filtered_movies.sort_values(by='average_rating', ascending=False).head(topN)
    filtered_movies['Sno.'] = range(1, len(filtered_movies) + 1)
    output = filtered_movies[['Sno.', 'title', 'average_rating', 'num_reviews']]
    
    display(output)

def on_content_recommendation_clicked(b):
    movie_title = title.value
    topN = num_recommendations_2.value
    try:
        output = recommendation_genre(movies3, cosine_sim, movie_title, topN)
        display(output)
    except ValueError as e:
        print(e)

# Attach callbacks to buttons
b1.on_click(on_popularity_recommendation_clicked)
b2.on_click(on_content_recommendation_clicked)
```
